## Check - scikit-learn

```python
from sklearn.neighbors import NearestNeighbors
import pandas as pd
samples = [[2,0], [5,0], [2,3], [6,4]]
points = [[3, 6.8], [4.5, 3], [ 3.3, 1.3]]

for metric in ['euclidean', 'manhattan', 'chebyshev']:    
    neigh = NearestNeighbors(n_neighbors=1, metric=metric)
    neigh.fit(samples) 
    dists, nn = neigh.kneighbors(points) 
    res = pd.DataFrame(np.column_stack((dists, nn)))
    res.columns = ['Distance', 'NN']
    res['NN'] = res['NN'].apply(lambda n: (samples[int(n)][0], samples[int(n)][1]))
    res['Point'] = points
    cols = res.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    res = res[cols]
    print('Metric: ', metric)
    print(res)
    print()
```

## Nearest Neighbors Classification

```python
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

n_neighbors = 15

# import some data to play with
iris = datasets.load_iris()

# we only take the first two features. We could avoid this ugly
# slicing by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target

h = .02  # step size in the mesh

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

for metric in ['euclidean', 'manhattan', 'chebyshev']:
    
    for weights in ['uniform', 'distance']:
        # we create an instance of Neighbours Classifier and fit the data.
        clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights, metric=metric)
        clf.fit(X, y)

        # Plot the decision boundary. For that, we will assign a color to each
        # point in the mesh [x_min, x_max]x[y_min, y_max].
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        plt.figure()
        plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

        # Plot also the training points
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                    edgecolor='k', s=20)
        plt.xlim(xx.min(), xx.max())
        plt.ylim(yy.min(), yy.max())
        plt.title("Metric - %s (k = %i, weights = '%s')"
                  % (metric, n_neighbors, weights))
        

plt.show()
```

## Mean distance

```python
one_color_points = 700
mean1 = [0, 0]
cov1 = [[1, 0], [0, 1]]  # diagonal covariance
x1, y1 = np.random.multivariate_normal(mean1, cov1, one_color_points).T
mean2 = [3, 3]
cov2 = [[1, 0], [0, 1]]  # diagonal covariance
x2, y2 = np.random.multivariate_normal(mean2, cov2, one_color_points).T
mean3 = [4, 0]
cov3 = [[1, 0], [0, 1]]  # diagonal covariance
x3, y3 = np.random.multivariate_normal(mean3, cov3, one_color_points).T

n = 5
points_x = np.random.choice(500, size=(n, 1))/100
points_y = np.random.choice(400, size=(n, 1))/100

plt.figure(figsize=(10,10))
plt.plot(x1, y1, 'bx')
plt.plot(x2, y2, 'yx')
plt.plot(x3, y3, 'gx')
plt.plot(points_x, points_y, 'ro')
plt.axis('equal')



plt.grid(True)

plt.show()
```

```python
x = np.array(points_x)
y = np.array(points_y)
points = np.column_stack((x, y))

data1 = np.column_stack((x1, y1))
data2 = np.column_stack((x2, y2))
data3 = np.column_stack((x3, y3))

samples = np.concatenate((data1, data2))
samples = np.concatenate((samples, data3))
print(samples.shape)
n_neighbors = 100
neigh = NearestNeighbors(n_neighbors=n_neighbors, metric='euclidean')
neigh.fit(samples) 
dists, nn = neigh.kneighbors(points) 

radius = []
for d in dists:
    radius.append(max(d))

print(radius)
r_mean = np.mean(radius)
print(r_mean)
r_mean_10 = r_mean / 10
print(r_mean_10)
```
