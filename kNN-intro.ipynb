```python
import numpy as np
import matplotlib.pyplot as plt

# evenly sampled time at 200ms intervals
#([1,2,3,4], [1,4,9,16], 'ro')


# red dashes, blue squares and green triangles
#plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.plot([1, 1.1, 0.95, 1.1, 1.05], [0,0.1,0.05, 0.1, 0.1],  'ro')
plt.plot([1,1.1,1.05, 1.1, 1.1], [1, 1.1, 0.95, 1.1, 1.05], 'g^')
plt.axis([0, 1.2, -0.1, 1.2])
plt.grid(True)
plt.xlabel('HACK')
plt.ylabel('COMPUTER')
plt.show()
```
